On peut créer un rdd à partir d'énormément de sources :

- Un fichier csv
- Un fichier json (séparé par des retours à la ligne)

Au moment où on fait parallelize, le rdd n'est pas encore véritablement créé.

### Les arbres de Paris d'hier au format CSV

1. Trouver de quoi lire le fichier csv en mode texte classique
2. Spark a-t-il effectué une transformation pour transformer les lignes du fichiers en "cellule" ?

Nous allons effectuer une transformation basique et naïve.

2. (bis) Combien y-a-t-il de lignes dans ce fichier ? De caractères valant ; ?
3. Lire la première ligne du fichier et extraire les noms des colonnes
4. Faire en sorte d'avoir un nouvel rdd avec non pas un tableau de lignes mais un tableau de dictionnaires indexés par les noms des colonnes.
5. Spark (ou Python) considère-t-il une colonne comme des entiers ou des float ? Pourquoi ?
6. Faire en sorte de traiter les entiers comme des entiers, les float comme des float (notamment hauteur...)

```python
rdd = sc.textFile("les-arbres.csv")
print("Nombre de lignes",rdd.count())
print(rdd.map(lambda ligne: ligne.count(";")).sum())

firstRow = rdd.first()
print("Première ligne",firstRow)
headers = firstRow.split(";")

def filterFirstRow(idx,iter):
    if idx == 0:
        next(iter)
    return iter

newRdd = rdd.mapPartitionsWithIndex(filterFirstRow)

#"JEROME;CHAABAN;18" ==> [“JEROME,“CHAABAN”,“18"] ==> {nom:CHAABAN,prenom:JEROME,age:18}

def fromLineToDict(line):
    splittedLine = line.split(";")
    toBeReturn = {header:splittedLine[headerNumber] for headerNumber,header in enumerate(headers)}
    toBeReturn["HAUTEUR (m)"] = float(toBeReturn["HAUTEUR (m)"])
    return toBeReturn

rddDict = newRdd.map(fromLineToDict)
print(rddDict.take(3))

print(rddDict.filter(lambda x: x["HAUTEUR (m)"] >= 10).count())
```
